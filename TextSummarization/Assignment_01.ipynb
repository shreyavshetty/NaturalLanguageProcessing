```python
from os import path
import nltk
from wordcloud import WordCloud
from nltk.tokenize import RegexpTokenizer
nltk.download('punkt')
```

```python
nltk.download()
```

```python
#sentence tokenanization
f = open('result1.txt', 'r')
text=f.read()
sents = nltk.sent_tokenize(text)
print(sents)
```

```python
#word tokenize every sentence 
#convert every word to its lower case and eliminate punctuations
words = []
w = []
for sent in sents:
    tokenizer = RegexpTokenizer(r'\w+')
    w = tokenizer.tokenize(sent)
    for word in w:
        words.append(word)
words = [w.lower() for w in words]
words
```

```python
#stop_words conatins a list of all stop words
#appended additional stop words to that list
stop_words = nltk.corpus.stopwords.words('english')
stop_words.append('(')
stop_words.append(')')
stop_words.append('.')
stop_words.append(',')
stop_words.append('I')
stop_words.append("'s")
stop_words.append('...')
stop_words.append('?')
stop_words.append('!')
set(stop_words)
```

```python
#remove all stop words from the list of words
filtered_word = [w for w in words if w not in stop_words]
print(filtered_word)
```

```python
#frequecy of all words in the given corpus
words_fd = nltk.FreqDist(filtered_word)
words_fd
```

```python
#list the top 30 words
d = words_fd.most_common(30)
print(d)
value = [e[0] for e in d ]
print(value)
```

```python
#calculate score for every sentence
sent_score = {}
for sent in sents:
    score = 0
    tokenizer = RegexpTokenizer(r'\w+')
    s = tokenizer.tokenize(sent)
    s = [w.lower() for w in s]
    for word in s:
        if word in filtered_word:
            score = score + words_fd[word]
    sent_score[sent] = score
print(sent_score)
```

```python
#sort based on sentence score and display top 10 sentences
#summarization
import operator
sorted_x = sorted(sent_score.items(), key=operator.itemgetter(1))
#print(sorted_x[::-1][:10])
top_10 = sorted_x[::-1][:10]
summary = ''
for i in top_10:
    summary = summary + i[0]
print(summary)
```

```python
#pos tag on each word
d = nltk.pos_tag(filtered_word)
noun = []
adjective = []
#create a seperate list of nouns and adjectives
for w in d:
    if w[1] == 'NN' or w[1] == 'NNP'or w[1] == 'NNS'  or w[1] == 'NNPS':
        noun.append(w[0])
    if w[1] == 'JJ' or w[1] == 'JJR'or w[1] == 'JJS':
        adjective.append(w[0])
```

```python
#create wordCloud
max_occured = ''
for i in value:
    max_occured = max_occured + i + ' '
nn = ''
for i in noun:
    nn = nn + i + ' '
adj = ''
for i in adjective:
    adj = adj + i + ' '
wordcloud = WordCloud().generate(max_occured)
wordcloud.to_file("/home/shreyavshetty/Desktop/max_occured.png")
wordcloud = WordCloud().generate(nn)
wordcloud.to_file("/home/shreyavshetty/Desktop/noun.png")
wordcloud = WordCloud().generate(adj)
wordcloud.to_file("/home/shreyavshetty/Desktop/adjective.png")
```

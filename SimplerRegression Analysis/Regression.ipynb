```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from pandas.plotting import scatter_matrix
#Download the dataset
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/parkinsons/telemonitoring/parkinsons_updrs.data"
data = pd.read_csv(url)
```

```python
#Load the dataset into data frame
df = pd.DataFrame(data)
print(df)
#statistical information about the data (various variables)
df.describe()
```

```python
'''
Total_UPDRS column is chosen to be your dependent variable. Drop the other candidate dependent
variable motor_UPDRS.
'''
plt.plot(df['motor_UPDRS'])
plt.show()
```

```python
plt.plot(df['total_UPDRS'])
plt.show()
```

```python
#plot same for motor_UPDRS and total_UPDRS
#Drop the other candidate dependent variable motor_UPDRS
df = df.drop(['motor_UPDRS'], axis=1)
```

```python
#data distributions with box plots 
l = df.columns
for i in l:
    plt.boxplot(df[i])
    print(i)
    plt.show()
```

```python
#data distributions with histograms
for i in l:
    plt.hist(df[i])
    print(i)
    plt.show()
```

```python
#skewed variables
skewed = ["Jitter(%)","Jitter(Abs)","Jitter:RAP","Jitter:PPQ5","Jitter:DDP","Shimmer","Shimmer(dB)","Shimmer:APQ3","Shimmer:APQ5","Shimmer:APQ11","Shimmer:DDA","NHR","DFA","PPE"]
# apply log(x) element-wise to a subset of columns
for column in df:
    if column in skewed:
        df[column] = np.log(df[column])
print(df)
```

```python
#Create a scatter matrix
scatter_matrix(df, alpha=0.2, figsize=(21,30), diagonal='kde')
plt.show()
print(df.columns)
```

```python
'''
>>Jitter(%) is highly colinear with Jitter(abs),Jitter:RAP,Jitter:PPQ5,Jitter:DDP
>>Jitter(Abs) is highly colinear with Jitter(%),Jitter:RAP,Jitter:PPQ5,Jitter:DDP
>>Jitter:PPQ5 is highly colinear with Jitter(%),Jitter(Abs),Jitter:RAP,Jitter:DDP
>>Jitter:DDP is highly colinear with Jitter(%),Jitter(Abs),Jitter:RAP,Jitter:PPQ5
>>Shimmer is highly colinear with 'Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:APQ11',
       'Shimmer:DDA'
>>Shimmer(dB) is highly colinear with Shimmer, 'Shimmer:APQ3', 'Shimmer:APQ5', 'Shimmer:APQ11',
       'Shimmer:DDA'
>>Shimmer:APQ3 is highly colinear with Shimmer, 'Shimmer(dB)', 'Shimmer:APQ5', 'Shimmer:APQ11',
       'Shimmer:DDA'
>>Shimmer:APQ5 is highly colinear with Shimmer, 'Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ11',
       'Shimmer:DDA'
>>Shimmer:APQ11 is highly colinear with Shimmer, 'Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'Shimmer:DDA'
>>Shimmer:DDA is highly colinear with Shimmer, 'Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5'
'''
```

```python
'''
So, the independent variables are subject,age,sex,test_time,total_UPDRS,Jitter(%),Shimmer,NHR,HNR,RPDE,DFA,PPE
'''
```

```python
dependent = ['Jitter(Abs)', 'Jitter:RAP', 'Jitter:PPQ5', 'Jitter:DDP','Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5','Shimmer:APQ11','Shimmer:DDA']
```

```python
#seaborn library to perform regression analysis between age and shimmer
for columns in df:
    if columns not in dependent and columns != 'total_UPDRS':
        fig = sns.regplot(data = df, x = columns, y = 'total_UPDRS')
        plt.show()
```
